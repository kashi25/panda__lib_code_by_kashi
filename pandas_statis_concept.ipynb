```python
# dickey-fuller test in stationarity
```

```python
data = pd.read_csv('Course+DataSet.csv', index_col = 'DATE', parse_dates = True)
```

```python
data.head()
```

```python
from statsmodels.tsa.stattools import adfuller
import pandas as pd

def adf_test(series):
    result = adfuller(series, autolag="aic")  # Corrected "ACI" to "aic"
    labels = ["ADF Test statistics", "p-value", "#lags used", "# observations"]
    out = pd.Series(result[0:4], index=labels)
    print(out.to_string())
    if result[1] <= 0.05:
        print("Reject the null hypothesis. Data has no unit root and stationary")
    else:
        print("Fail to reject the null hypothesis. Data has a unit root and is non-stationary")

adf_test(data["Price"])
```

```python
import pandas as pd
```

```python
data = pd.read_csv('Course+DataSet.csv', index_col = 'DATE', parse_dates = True)
```

```python
data.head()
```

```python
autocorrelation_lag1 = data['Price'].autocorr(lag = 1)
print("one month lag", autocorrelation_lag1)
```

```python
autocorrelation_lag2 = data['Price'].autocorr(lag = 2)
print("two months lag", autocorrelation_lag2)
```

```python
autocorrelation_lag3 = data['Price'].autocorr(lag = 3)
print("three months lag", autocorrelation_lag3)
```

```python
autocorrelation_lag6 = data['Price'].autocorr(lag = 6)
print("six months lag", autocorrelation_lag6)
```

```python
autocorrelation_lag9 = data['Price'].autocorr(lag = 9)
print("nine months lag", autocorrelation_lag9)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Course+DataSet.csv", index_col = 'DATE', parse_dates = True)
```
